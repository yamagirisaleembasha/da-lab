```python
import pandas as pd
import numpy as np
```

```python
nur=pd.read_csv('National_Universities_Rankings.csv')
nur
```

```python
#1
nur[nur['Undergrad Enrollment'].notnull()]['Name']
```

```python
#2
state=nur['Location'].str.split(', ',expand=True)
nur['state']=state[1]
nur['city']=state[0]
nur
```

```python
#3
list(nur.groupby(['state','city','Name']).groups.keys())
```

```python
#4
print(len(list(nur[nur['Rank']<50].groupby(['state','Rank','Name']).groups)))
list(nur[nur['Rank']<50].groupby(['state','Rank','Name']).groups)
```

```python
#5
nur[nur['Tuition and fees'].notnull() & nur['In-state'].notnull()]['Name'] 
```

```python
#6
fee_out = nur['Tuition and fees'].str.replace('\W', '' ,regex=True).astype(int)
fee_In = nur['In-state'].fillna('0')
fee_In = fee_In.str.replace('\W', '' ,regex=True).astype(int)
nur[abs(fee_out-fee_In)<=5000]
```

```python
#7.	List the universities having tuition fee>15000$ and rank between 120 to 170
nur[(nur['Tuition and fees'].str.replace('\W', '' ,regex=True).astype(int)>15000) & (nur['Rank']> 120) & (nur['Rank']<170)]
```

```python
# 8.	Find the campuses of universities located in different cities (multiple cities)
nur[['Name', 'city']] = nur['Name'].str.split('--', expand=True)
print(nur[['Name', 'city']])
nur[nur["Name"].duplicated()==True]["Name"]
```

```python
#9
a=nur['In-state'].fillna('$0')
a=a.str.replace('\W', '' ,regex=True).astype(int)
print(nur[(nur['Tuition and fees'].str.replace('\W', '' ,regex=True).astype(int))>a]['state'])
print('min max in_state:' ,min(a),max(a))
print('min max out_state:',min(nur['Tuition and fees'].str.replace('\W', '' ,regex=True).astype(int)),max(nur['Tuition and fees'].str.replace('\W', '' ,regex=True).astype(int)))
```

```python
#10.	Find the cities locating top 100 universities
nur[nur['Rank']<100]['city']
```

```python
#11.	Find universities with least no of undergraduate students.
nur[nur[nur['Undergrad Enrollment'].notnull()]['Undergrad Enrollment']==min(nur[nur['Undergrad Enrollment'].notnull()]['Undergrad Enrollment'])]['Name']
```

```python
#12.	Identifying correlations between enrollment numbers and university rank
import scipy.stats
a=scipy.stats.pearsonr(pd.Series(nur['Undergrad Enrollment'].str.replace('\W', '' ,regex=True).astype(int)), 
                     pd.Series(nur['Rank']))[0]
print('correlations between enrollment numbers and university rank : %.4f' % a)
```
