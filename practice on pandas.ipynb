```python
import pandas as pd
```

 # Data exploration

```python
data ={'Name':['Jai', 'Princi', 'Gaurav','Anuj', 'Ravi', 'Natasha', 'Riya'],
        'Age': [17, 'NaN', 18, 17, 18, 17, 17],
        'Gender': ['M', 'F', 'M', 'M', 'M', 'F', 'F'],
         'Marks': [90, 76, 'NaN', 74, 65, 'NaN', 71]}
df = pd.DataFrame(data)
```

```python
df
```

# Dealing with missing values

```python
for i in data.keys():
    if 'NaN' in list(df[i]):
        s=0
        c=0
        for j in list(df[i]):
            if j=='NaN':
                pass
            else:
                s+=j
                c+=1
        for j in list(df[i]):
            if j=='NaN':
                df[i][list(df[i]).index(j)]=s/c
```

```python
df
```

# Reshaping data

```python
for i in list(df['Gender']):
    if i=='M':
        df['Gender'][list(df['Gender']).index(i)]=0
    else:
        df['Gender'][list(df['Gender']).index(i)]=1
```

```python
df
```

# Filtering data

```python
df =df[df['Marks'] >= 75] 
df=df.drop(['Age'], axis=1)
```
